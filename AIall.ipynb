```python
#BFS
from collections import deque
graph={}
n=int(input("Enter the no of nodes"))

for i in range(n):
    node=input("Enter the nodes")
    neighbors=input("Enter the neighbor").split()

    graph[node]=neighbors

start=input("Enter the start")

visited=set()
queue=deque([start])

while queue:
    node=queue.popleft()

    if node not in visited:
        print(node, end="")
        visited.add(node)

    for i in graph[node]:
        queue.append(i)
        
```
